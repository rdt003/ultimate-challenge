```python
import pandas as pd
import numpy as np
import json
import requests
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
```

# <div class="alert alert-warning"> Part 1 â€‘ Exploratory data analysis </div>

```python
# Load in the logics.json file
logins = pd.read_json('logins.json')
logins.head()
```

```python
# Set index to time of login
logins = logins.set_index('login_time')
# Create a login column
logins['logins'] = 1
# Aggregate logins in 15m intervals
logins = logins.resample('15T').count()
logins['day'] = list(logins.reset_index().login_time.apply(lambda x: x.day))
logins['weekday'] = list(logins.reset_index().login_time.apply(lambda x: x.weekday()))
logins['hour'] = list(logins.reset_index().login_time.apply(lambda x: x.hour))
logins.head()
```

```python
# Plot the login_time and logins
plt.figure(figsize=(15,15))
plt.plot(logins.index, logins.logins);
```

> The plot does not do a good job of displaying patterns of demand, so I'll have to take a more direct approach.

> Lets start by examining the median logins over the span of a month to determine whether or not certain times of the month had higher login rates. Ex: start/middle/end of month had increased number of logins.

```python
# View the median logins by 'day'
# 'day' refers to the day's numerical date
logins.groupby('day').logins.median()
```

> There doesn't appear to be a difference between the start, middle, or end of the month.

> Lets try finding a pattern across the weekdays.

```python
# View the median logins by 'weekday'
# 'weekday' refers to the day of the week
  # using 0-6 to represent Monday-Sunday
logins.groupby(['weekday']).logins.median()
```

> The most logins typically occur on Friday, Saturday, and Sunday. 

> Now that we know this, we can examine the daily cycles to determine which hours result in the most frequent logins.

```python
temp = logins.groupby(['weekday', 'hour']).logins.median()
temp.unstack(level=0)
```

> On the weekdays (Monday - Friday), most logins typically occur between the hours of 10AM-2PM and 8PM-2AM.
<br> Friday is special in that it has a high amount of logins between the hours of 10AM-11PM.

> On the weekends (Saturday-Sunday), most logins typically occur between the hours of 12AM-6AM and 12PM-11PM.

# <div class="alert alert-warning"> Part 3 â€‘ Predictive modeling </div>

```python
# Open the file
with open('ultimate_data_challenge.json') as json_file:      
    data = json_file.readlines()
    data = list(map(json.loads, data)) 

# Create DataFrame
user_data = pd.DataFrame(data[0])

# Check for datatype errors
print(user_data.info())
print('\n')
# Check for null values
print('Null Values:')
print(user_data.isna().any())

# View first 5 rows of data
user_data.head()
```

```python
# Determine max value for avg_rating columns
print('Max avg rating by driver:', user_data.avg_rating_by_driver.max())
print('Max avg rating of driver:', user_data.avg_rating_of_driver.max())
print('\n')

# Determine included city values
print('Included cities:', user_data.city.unique())
print('\n')

# Determine included phones
print('Included phones:', user_data.phone.unique())
```

```python
# Fill null values for avg_rating columns with 5.0
avg_rating_columns = ['avg_rating_by_driver', 'avg_rating_of_driver']
user_data[avg_rating_columns] = user_data[avg_rating_columns].fillna(5.0)
# Fill null values for phone column with 'other'
user_data['phone'] = user_data['phone'].fillna('Other')

# Determine the cuttoff date for retained users
cutoff_date = pd.to_datetime(user_data.last_trip_date).max() - datetime.timedelta(30)
# Create 'retained' users column
user_data['retained'] = pd.to_datetime(user_data.last_trip_date) >= cutoff_date

# Create boolean column for each city and each phone
user_data[['astapor', 'kings_landing', 'winterfell']] =  pd.get_dummies(user_data.city)
user_data[['android', 'iphone', 'other']] = pd.get_dummies(user_data.phone)
# Drop city and phone columns
user_data = user_data.drop(['city','phone'], axis=1)

# Convert 'ultimate_black_user' column to 0/1
user_data['ultimate_black_user'] = user_data.ultimate_black_user * 1

# Drop date columns
user_data = user_data.drop(['last_trip_date', 'signup_date'], axis=1)

# Instantiate feature columns
features = user_data.drop(['retained'], axis=1).columns
```

```python
from sklearn.model_selection import GridSearchCV

# Split feature data and target data
X = user_data[features]
y = user_data['retained']

# Set parameter grid
param_grid = {'n_estimators': [1,5,10,25,50,100], 'max_features': np.arange(1,15)} 
rfc_cv = GridSearchCV(RandomForestClassifier(), param_grid, cv=5)
rfc_cv.fit(X, y)
print('Best parameters:', rfc_cv.best_params_)
print('Best score:', rfc_cv.best_score_)
```

```python
# Create train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.40, random_state=42)

# Standardize the features
temp = StandardScaler().fit(X_train)
X_train = temp.transform(X_train)
X_test = temp.transform(X_test)

# Set up our RandomForestClassifier and fit to data
clf = RandomForestClassifier(n_estimators=50, max_features=13, random_state=42)
clf.fit(X_train, y_train)

# Make predictions
predictions = clf.predict(X_train)

# Print score
print('Training Accuracy:', clf.score(X_train, y_train))
print('Test Accuracy:', clf.score(X_test, y_test))
```
